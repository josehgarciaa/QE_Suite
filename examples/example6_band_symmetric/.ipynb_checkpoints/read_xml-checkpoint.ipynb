```python
import qe_suite.format as f
import numpy as np
import xml.etree.ElementTree as ET
tree = ET.parse('qe_suite/qe_suite.xml')
root = tree.getroot()
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
class constant:
    Ry2eV= 13.605684958731;
c =constant();

def band_structure(xml_dir="qe_suite/qe_suite.xml "):
    #Parse xml
    tree = ET.parse(xml_dir)
    root = tree.getroot()
    bs   = root.find("output/band_structure");
    EF   = float( root.find("output/band_structure/fermi_energy").text );
    #
    kpoints= [(x.find("k_point").text).split(" ") for x in bs.iter("ks_energies")]
    eigvals= [f.remove_empty((eigv.find("eigenvalues").text.replace("\n","")).split(" ")) for eigv in bs.iter("ks_energies")]
    eigvals= (np.array(eigvals, dtype=float).T - EF)*c.Ry2eV;
    kpoints= np.array(kpoints, dtype=float);
    kpath =  np.insert( np.cumsum( np.linalg.norm( np.diff( kpoints, axis=0 ), axis=1 ) ),0,0)
    
    import matplotlib.pyplot as plt
    for eig in eigvals:
        plt.plot(kpath,eig)
    
```

```python
band_structure('qe_suite/qe_suite.xml')
```

```python
import matplotlib.pyplot as plt
for eig in eigvals:
    plt.plot(kpath,eig)
#plt.ylim(-3,3)
```
